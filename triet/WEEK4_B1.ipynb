# Phạm trù (học thuyết) hình thái kinh tế xã hội
1. Phạm trù hình thái kinh tế xã hội và cấu trúc hình thái kinh tế xã hội.
    a. Phạm trù hình thái kinh tế xã hội.
    b. Cấu trúc của hình thái kinh tế xã hội.
        - Phải hiểu về đối tượng.
        - Hiểu về mối quan hệ (Lưu ý).

2. Biện chứng giữa lực lượng sản xuất và quan hệ sản xuất.
    a. Lực lượng sản xuất: 
    - Con người là lực lượng sản xuất.
    - Những gì thuộc giới tự nhiên tham gia vào sản xuất được gọi là tư liệu sản xuất: Tư liệu sản xuất được chia làm 2 loại là tư liệu lao động và đối tượng lao động.
    - Tư liệu lao động: Công cụ lao động và phương tiện lao động.

Note:
    - Người lao động là yếu tố giữ vai trò quyết định.
    - Công cụ lao động là yếu tố cực kỳ quan trọng.
    - PTSX: Phương tiện sản xuất.
