# Vấn đề cơ bản của triết học là gì?

2. Khái lược triết học phương đông.
a. Triết học ấn độ:
    1. Chịu ảnh hưởng lớn của tư tưởng tôn giáo.
    2. Mang tính hướng nội, nặng về tâm linh. Tập trung vào vấn đề giải thoát kiếp sau bằng đời sống đức hạnh.
    3. Hệ thống tư tưởng.
    4. Thể hiện khái quát và tư duy biện chứng.
b. Bối cảnh xã hội:
    - Nền kinh tế trong xã hội ấn độ cổ đại là nền kinh tế kết cấu theo mô hình công xã nôn thôn.
    - Khoảng thế kỷ 15 TCN: Aryan(Arya): Người du mục độ bào ấn độ. Xã hộ ấn độ phân chia làm 4 đẳng cấp:
        1. Đảng cấp thần quyền (Bà la Môn).
        2. Đẳng cấp thế quyền  (Vua chúa).
        3. Đẳng cấp dân Tự do (Thủ công, buôn bán).
        4. Đẳng cấp Nô lệ (đông nhất).
    -> Theo kinh Vedas 4 đẳng cấp trong xã hội là hiện thân của Linh hồn cũ trụ: Kinh Vedas giải thích tại sao lại phân làm 4 đẳng cấp. Đặc trung cho 4 bộ phận khác nhau của thần thánh.
    -> Càng ngày sự bóc lột của những đẳng cấp thống trị càng tàn bạo, ranh giới giữa các đẳng cấp ngày càng trở nên nghiệt ngã. Tạo nên cuộc sống hết sức ngột ngạt, u ám => Khát vọng được giai thoát, có cuộc sống bình đẳng, bác ái bao trùm khắp đất nước ấn độ.
    -> Ngày nay nói đến triết học Ấn Độ là nói đến triết học giải thoát.

-> Câu hỏi: Đời sống tinh thần của con người nói chung và vấn đề triết học nói riêng không phải do con người quyết định mà do môi trường quyết định.

-> Công cụ câm: phương tiện, công cụ bán câm: trâu bò và nô lệ.

c. Các trường phái:
Ấn độ có 9 phái lớn:
- Chính thống: tuân theo Vedas (tất cả mọi thứ không thể thay đổi được vì nó là hiện thân của linh hồn vũ trụ) 3 học thuyết:
    + Samkhya, Vaisesika, Nyaya, Yoga, Mymansa, Vedanta.
- Tà giáo (Phi chính thống) 3 học thuyết:
    + Lokayata, Buddhism, Jaina.

d. Triết học phật giáo (Buddhism):
Sự ra đời: Là trào lưu triết học tôn giáo ra đời vào thế kỷ 6 TCN: Người sáng lập là Tất Đạt Đa: 6 năm tầm sư học đạo, 49 ngày ngồi dưới gộc bồ đề (tức giác ngộ), 45 năm thuyết pháp.
- Kinh điển (sách gốc (sách rõ ràng nhất) để tìm hiểu về phận giáo):
    + Tam tạng (3 bộ sách, 3 giỏ):
        1. Tạng kinh (Đại tạng kinh).
        2. Tạng luận (ghi những lời bình luận của Phật tử).
        3. Tạng luật: Ghi những giới (những điều không được làm) và luật (là những điều con người phải làm theo).

-> Nhập diệp (mất):

- Một số nộ dung cơ bản của triết học Phật giáo
    + Vũ trụ quan (tự bản thân tồn tại, không có điểm khởi đầu và kết thúc): Vũ trụ vô thuỷ vô chung (không có kiểm khởi đàu và kết thúc), vạn pháp vô thường (luân biến đổi), vô ngã (không có bản thể), giả tướng, vận hành theo luật nhân quả.
    + Nhân sinh quan (quan niệm  về cuộc sống): Con người là duyên hợp của ngũ uẩn (sắc uẩn, thụ uẩn, tưởng uẩn, hành uẩn, thức uẩn):
        - Ngũ uẩn chia thành phần danh và phần sắc.
        - Con người là thực thể vô thương, vô ngã, giả tướng: tuỳ theo nghiệp báo mà vận hành qua các kiếp theo 6 cõi tái sinh (lục đạo luân hồi):
            + Cõi người
            + Trời
            + Atula
            + Quỷ
            + Súc sinh
            + Địa ngục
        - Con người có thể tự mình gỉai thoát được nỗi khổ bằng các hiểu và thực hiện được tứ diệu đế thì sẽ giải thoát được:
            + Tứ diệu đế (4 nền tẳng kỳ diệu) bao gốm:
                1. Khổ đế: Là quan điểm của Phật giáo về nỗi khổ của cuộc đời (8 khổ cơ bản):
                    - Sinh khổ
                    - Lão khổ
                    - Bệnh khổ
                    - Tử khổ
                    - Ái biệt ly khổ
                    - Oán hội khổ
                    - Cầu bất đắc khổ
                    - Ngũ uẩn khổ (5 giác quan)
                -> 4 đầu sinh lý, 3 tiếp tâp lý, cuối cả hai (sinh lý + tâm lý).
                2. Tập đế: Là nhũng nguyên nhân gây ra nỗi khổ:
                    - Thập nhị nhân duyên (12 nguyên nhân ràng buộc gây ra nỗi khổ) => 3 nguyên nhân cơ bản (Tham, Sân (sân hận), Si (si mê)): khi lòng tham không được đáp ứng => sân hận => si mê. 
                    - Căn nguyên của mọi căn nguyên: vô minh (tố tăm về mặt trí tuệ): không biết mình là ai.
                3. Diệt đế.
                4. Đạo đế.
