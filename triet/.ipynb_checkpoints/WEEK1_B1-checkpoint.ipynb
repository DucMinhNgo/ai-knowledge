# Mục đích môn học:

# Yêu cầu:
1. Yêu cầu về mặt học thuật:
- Chủ động học tập: nghe ghi.
- Hoàn toàn có quyền nghi ngờ: hoài nghi khoa học, đề xuất ý kiến.
- Vận dụng thế quan: phát hiện giải quyết những vấn đề thuộc chuyên ngành và cuộc sống đặt ra.
2. Yêu cầu về mặt giảng đường
- Không được sử dụng điện thoại trong giờ học.
- Không được làm ảnh hưởng đến sự tập trung của lớp học.
- Không nên đi muộn.

# Cấu trúc chương trình
1. Khái luận về triết học.
2. Triết học Mac-Lenin.
3. Mối quan hệ giữa triết học và khoa học.
4. Vai trò của công nghệ trong sự phát triển.

# Giáo trình

# Đánh giá kết quả:
+ Giữa kỳ + tiểu luận 40%.
+ Thi hết môn: 60%.

# Chương 1: Khái niệm về triết học
1. Triết học và vấn đề cơ bản của triết học: Ra đời gần như đồng thời ở cả phương đông (quốc gia dọc theo bờ sông Nin, sông ấn, sông Hoàng:Ai cập, Ấn độ, Trung quốc, Lưỡng hà) lẫn phương tây (Hy lạp, la mã) vào khoảng thế kỷ thứ 8 tcn.
- Chỉ tìm hiểu hai quốc gia đại diện cho:
    + Phương Đông (Trung Quốc): Tay (), Búa, Miệng (Khẩu): Con cáo vào trùm nho, con nai và thợ săn, mổ xẻ để hiểu được bản chất của vấn đề.
    + Phương Tây (Hy lạp): Philosophia (yêu mến sự thông thái).

-> Cách diễn đạt bằng nhiều cách phải tôn trọng lịch sử và bản chất của vấn đề.

- Định nghĩa triết học: là hệ thống lý luận chung nhất của con người và thế giới (vũ trụ) về bản chất con người và vị trí của con người trong thế giới ấy. Đặc trưng nhất của tri thức.
    + Vế này cho thấy tri thức triết học là tri thức phải có:
        - Tính hệ thống (tri thức trước phải làm nền cho tri thức sau).
        - Tính lý luận.
        - Tính khái quát.
    
    + Vế sau đề cập đến nội dung mà triết học phản ánh:
        - Những cấn đề về thế giới.
        - Vấn đề về con người, xã hội.
        - Vị trí, vai trò của con người trong thế giới.

-> Vấn đề cụ thể chỉ là tư liệu: triết học đưa ra những giải thích rất hệ thống các lý luận mang tính khái quát(khái quát dưới dạng lý luận).
