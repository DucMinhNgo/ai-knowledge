# Lịch sử trung quốc
- Trung quốc thời Tam Đại đã xuất hiện đã hình thành thử tưởng: Âm Dương, Ngũ hành, Thượng đế: Triết học phát triển mạnh vào thời chu.
- Nhiều học thuyết ta đời: Bách gia chi tử:
- Khái quát lại chia làm 6 phái lớn (Lục gia): 5 phái đề cập về vấn đề xã hội, chỉ có Âm Dương gia đề cập đến vấn đề vũ trụ.
    + Nho gia.
    + Mặc gia.
    + Đạo gia.
    + Danh gia.
    + Âm Dương gia.
    + Pháp gia.

- Nho gia: (Sáng tạo Nho gia là Khổng Tử (Người thầy họ Khổng)): được Mạnh Tử và Tuân Tử từng bước bổ sung và phát triển.


- Hạ - Ân - Chu:
    + Tây Chu (Bình).
    + Đong Chu (Loạn).
