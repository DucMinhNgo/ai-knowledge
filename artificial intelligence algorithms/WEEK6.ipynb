# Adversarial search
- zero-sum games of perfect information (quan sát được toàn bộ môi trường).

- minimax algorithm có hai khuyết điểm chính:
    + đòi hỏi đối phương phải chơi tối ưu.
    + độ phức tạp quá lớn.

- Min cập nhật Beta, Max cập nhật Alpha

- Sử dụng phương pháp tỉa nhánh để giảm thời gian chạy của thuật toán.

- Tìm kiếm đối kháng.
