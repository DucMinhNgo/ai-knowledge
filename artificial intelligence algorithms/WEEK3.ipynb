# Breadth first search (tìm kiếm theo chiều rộng) (BFS).
queue

frontier set (tập đang duyệt)

- Tìm ra lời giải tối ưu nếu chi phí giữa các cạnh bằng nhau (nếu không thì sẽ không tối ưu).

priority queue

# Depth first search (DFS)
stack

Một số cải tiến (depth limited search) (DLS).

# Bidirectional
Tìm từ trạng thái đầu đến đích và từ dích đến đầu (thuật toán kết thúc khi hai chiều gặp nhau)
Chưa chắc tìm được đường đi ngắn nhất (nhiều trường hợp không thể đi ngược lại được).

Tìm được lời giải tốt nhưng chưa chắc là lời giải tối ưu.
