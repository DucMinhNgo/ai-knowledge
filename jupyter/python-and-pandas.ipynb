# Khởi tạo dữ liệu với:
- Dictionary
- List
- Mảng 2 chiều Numpy
- Files


```python
# Tạo pandas với dictionary
d = {'x': [1, 2, 3], 'y': np.array([2, 4, 8]), 'z': 100}
pd.DataFrame(d, index=[100, 200, 300], columns=['z', 'y', 'x'])
```

```python
# Tạo pandas với dữ liệu list
l = [
  {'x': 1, 'y': 2, 'z': 100},
  {'x': 2, 'y': 4, 'z': 100},
  {'x': 3, 'y': 8, 'z': 100}
]

pd.DataFrame(l)

# nest list
nl = [
  [1, 2, 100],
  [2, 4, 100],
  [3, 8, 100],
]

pd.DataFrame(nl, columns=['x', 'y', 'z'])
```

```python
# Tao dữ liệu pandas với numpy array
nl = np.array([
  [1, 2, 100],
  [2, 4, 100],
  [3, 8, 100],
])

df = pd.DataFrame(nl, columns=['x', 'y', 'z'])
df

# Việc không copy các giá trị dữ liệu có thể giúp bạn tiết kiệm thời gian, và sữ mạnh tính toán khi làm việc với tập dữ liệu lớn.
# Lưu dữ liêu vào file csv
df.to_csv('data.csv')
_df = pd.read_csv('data.csv', index_col=0)

_df
```
