# Các khái niệm cơ bản trong DL.
Perceptron (tế bào thần kinh)
Hàm kích hoạt (activation function): được sinh ra để bẻ gẫy sự tuyến tính. Các hàm này có thể hiểu đơn giản là bộ lọc để quyết định xem thông tin có được đi qua nơ ron hay không. Trong quá trình huấn luyện các hàm kích hoạt đóng vai trò quan trọng trong việc điều chỉnh độ dốc của đạo hàm. Một số hàm kích hoạt: sigmod, tanh, Relu. Hàm này thường là các hàm liên tục và khả vy (differentiable): tức là có đạo hàm tại mọi điểm trong miền xác định của nó.

- Sigmoid rất hữu ích để biến đổi bất kỳ giá trị nào thành xác suất (0, 1)
- Tanh là phiên bản thu nhỏ của hàm sigmoid đạo hàm của hàm này ổn hơn sigmod vì nó tránh được hiện tượng vanishing gradient. Cũng giống như sigmod hàm tanh được kích hoạt tại mọi thời điểm trong khoảng từ -1, 1 chính vì điều đó làm cho mạng của chúng có tham số khủng và rất nặng khi training để khắc phục vấn đề đó ta đưa ra hàm ReLU (Rectified Lear Unit)
- ReLU: loại bỏ bớt những tham số không cần thiết trong quá trình training. Giữ nguyên tất cả các giá trị lớn hơn 0, còn những giá trị nhỏ hơn 0 thì coi như bằng 0. Có một số biến thể là ELU và LeakReLU hoạt động cũng khá tốt.
