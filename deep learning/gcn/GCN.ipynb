- refine: lọc
- obtain: đạt được

- rationality: hợp lý

  - the most essential component of GCN
  - neighborhood aggregation for collaborative filtering
  - specifically: after associating each user (item) ưith an ID embedding
  - progagate the embedding on the user-item interaction graph to refine them
  => we combine the embedding learning at diffrent progagation layers with a weighted sum to obtain the final embedding for prediction.
