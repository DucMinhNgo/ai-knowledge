- An illustration of lightGCN model architecture. In LGC, only normalized sum of neighbor embeddings is performed towards next layer
- Other operations like seft-connection, feature transformation, nonlinear activation are all removed.
- In Layer Combination, we sum over the embedding at each layer to obtain the final representation.

# Light Graph Convolution (LGC).
=> Symmetric normalization: chuẩn hóa đối xứng has good performance
- essentially captures the same effects as self-connections. Thus, there is no need in LGC to include self-connections.

- only train parameter at Layer 0: all users, all items.
- It can be treated as a hyper-parameter to tuned manually
- The reason that we perform layer combination to get final representations are three fold:
  + with the increasing of the number, embedding will be over-smoothed.
  + the first layer enforces smoothness on users and items that have interactions, the second layer smoothes users(items) that have overlap on interactive items (users), higher layer capture higher order proximity (Lớp cao hơn nắm bắt mức đọ gần gũi cao hơn), combine will make representive more comprehensive
  + combining embedding at diffrent layer with weighted some weighted some capture, the effect of graph convolution with self-connections, an important trick in GCNs
