# Đặc trưng về dáng
Phải nắm được giải pháp truyền thống để hiểu được cách phát triển như thế nào

- compact (thể hiện sự trong chịa)
- Tỉ số giữa các trục chính

Bài tập 1: Cho 1 vùng tìm trục chính (tìm tâm dựa trên vector riêng giá trị riêng của ma trận hiệp phương sai của mỗi điểm trong vùng với giá trị tâm).
Bài tập 2: vì sao có tính compact tồi mà người ta vẫn phải tìm tỉ số giữa các trục chính (đưa lên phản ví dự cùng giá trị compact mà tỉ số giữa các trục chính vô để chứng minh, topo hình học).

- Bao lồi là tập lỗi bé nhất bao đối tượng (viết giải thuật tính bao lồi)
