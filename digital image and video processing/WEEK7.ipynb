# Phân đoạn ảnh (segmentation): là bài toán phân lớp ở từng điểm ảnh.
- Segmentation: là cluster phân nhóm các vùng có đặc trưng tương đồng về thị giác (không xuất ra id) boundary box (ranh giới).
- Object detection: xác định bound của đối tượng và mỗi đối tượng có 1 id riêng.
