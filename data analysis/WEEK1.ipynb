# Gradient based optimization
- Phương pháp vét cạn (Local Search):
  + Phương pháp gradient
    Gradient descent.
  + Không dùng gradient:
    Phương pháp Hill- climbing (leo đồi tham lam).
    Thuật toán annealing (thuật toán luyện thép).
  
- So sánh hai thuật toán dùng phương pháp gradient và phương pháp (free gradient):
  + free gradient không phù hợp cho không gian nhiều chiều.
