```python
# import library
import numpy as np
import matplotlib.pyplot as plt
```

```python
X0 = np.array([[2], [7], [9], [3], [10], [6], [1], [8]])

ones = np.ones_like(X0)

# Concat two matrix
X = np.concatenate((X0, ones), axis=1)
Y = np.array([[13], [35], [41], [19], [45], [28], [10], [55]])
```

```python
plt.scatter(X0, Y)
plt.show()
```

```python
def grad_cal(X, Y, theta_gd, m):
    """
    X: X's value
    Y: Y's value
    theta_gd: theta's value
    m: number of samples
    """    
    g = 1/m * X.T.dot(X.dot(theta_gd) - Y)
    return g.reshape(theta_gd.shape)
```

```python
def loss(X, Y, theta_gd, m):
    """
    X: X's value
    Y: Y's value
    theta_gd: theta's value
    m: number of samples
    """
    return 1/(2*m) * np.sum((X.dot(theta_gd) - Y)**2)
```

```python
# Normal Equation
# API np.linalg.inv() dng  tnh ma trn nghch o

theta = np.linalg.inv(X.T.dot(X)).dot(X.T.dot(Y))
theta
```

```python
def plot_polynomial(start, end, theta):
    # Generate x-values within the specified range
    x = np.linspace(start, end, 100)
    # Calculate the corresponding y-values using the polynomial function defined by theta
    y = np.polyval(theta, x)
    # Plot the polynomial curve
    plt.plot(x, y, color="orange")
```

```python
plt.scatter(X0, Y)
plot_polynomial(0, 10, theta)
plt.show()
```

```python
# S dng phng php gradient descent theo phn phi chun
# khi to theo phn phi chun
theta_gd = np.random.normal(size=2).reshape([2,1])
learning_rate = 0.02
theta_gd

def predict(X):
    return X.dot(theta_gd)

# Ch : lun phi ni 1 vo bn phi ca vector u vo ging nh ta  training.
X_new_1 = np.array([[40, 1]])
predict(X_new_1)
```

```python
m = 8
for i in range(10000):
    grad_value = grad_cal(X, Y, theta_gd, m)
    theta_gd = theta_gd - learning_rate*grad_value
    print ('=====')
    print(loss(X, Y, theta_gd, m))
    print (theta_gd)
```
