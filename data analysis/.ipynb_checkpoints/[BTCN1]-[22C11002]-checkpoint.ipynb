```python
# import library
import numpy as np
import matplotlib.pyplot as plt
```

```python
X0 = np.array([[7], [6.5], [6.5], [6.0], [6.0], [6.0], [5.5], [5.5], [5.0], [4.5]])
ones = np.ones_like(X0)

# Concat two matrix
X = np.concatenate((X0, ones), axis=1)
Y = np.array([[29], [32], [31], [34], [32], [35], [40], [43], [48], [50]])
print (X)
print (Y)
```

```python
# Cu 1
def plot_polynomial(start, end, theta):
    # Generate x-values within the specified range
    x = np.linspace(start, end, 100)
    # Calculate the corresponding y-values using the polynomial function defined by theta
    y = np.polyval(theta, x)
    # Plot the polynomial curve
    plt.plot(x, y, color="orange")

    
    
theta = np.linalg.inv(X.T.dot(X)).dot(X.T.dot(Y))
plt.scatter(X0, Y)
plot_polynomial(0, 10, theta)
plt.show()
```

```python
# predict
def predict_func(x):
    return x.dot(theta)

new_X = np.array([[4, 1]])

predict_Y = predict_func(new_X)

print (predict_Y)
#  ngha ca h s hi quy tuyn tnh, li sut ngn hng v tng vn u t t l ngh vi nhau, khi tng vn u t cng nh th li xut ngn hng cng ln
# nh gi mc  ph hp ca m hnh
```

```python
# Cu 2: Kim nh gi thit H s hi quy ca X trong hm hi quy tng th bng 0 vi mc  ngha 2% v nu  ngha ca kt qu?
import statsmodels.api as sm
import numpy as np

# To d liu mu
X = np.array([7, 6.5, 6.5, 6.0, 6.0, 6.0, 5.5, 5.5, 5.0, 4.5])
y = np.array([29, 32, 31, 34, 32, 35, 40, 43, 48, 50])

# Thm ct gi tr 1 vo ma trn X  tnh ton h s chn
X = sm.add_constant(X)

# To m hnh hi quy tuyn tnh
model = sm.OLS(y, X)

# Fit m hnh
results = model.fit()

results.summary()

# Kim nh gi thit
t_statistic = results.tvalues[1]  # Ly gi tr t-statistic ca h s X
p_value = results.pvalues[1]  # Ly gi tr p-value ca h s X

print("Gi tr t-statistic:", t_statistic)
print("Gi tr p-value:", p_value)
```
