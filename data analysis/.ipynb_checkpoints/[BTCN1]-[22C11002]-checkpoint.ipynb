```python
# import library
import numpy as np
import matplotlib.pyplot as plt
```

```python
X0 = np.array([[7], [6.5], [6.5], [6.0], [6.0], [6.0], [5.5], [5.5], [5.0], [4.5]])
ones = np.ones_like(X0)

# Concat two matrix
X = np.concatenate((X0, ones), axis=1)
Y = np.array([[29], [32], [31], [34], [32], [35], [40], [43], [48], [50]])
print (X)
print (Y)
```

```python
# Câu 1
def plot_polynomial(start, end, theta):
    # Generate x-values within the specified range
    x = np.linspace(start, end, 100)
    # Calculate the corresponding y-values using the polynomial function defined by theta
    y = np.polyval(theta, x)
    # Plot the polynomial curve
    plt.plot(x, y, color="orange")

    
    
theta = np.linalg.inv(X.T.dot(X)).dot(X.T.dot(Y))
plt.scatter(X0, Y)
plot_polynomial(0, 10, theta)
plt.show()
```

```python
# predict
def predict_func(x):
    return x.dot(theta)

new_X = np.array([[4, 1]])

predict_Y = predict_func(new_X)

print (predict_Y)
# Ý nghĩa của hệ số hồi quy tuyến tính, lãi suất ngân hàng và tổng vốn đầu tư tỷ lệ nghị với nhau, khi tổng vốn đầu tư càng nhỏ thì lãi xuất ngân hàng càng lớn
# Đánh giá mức độ phù hợp của mô hình
```

```python
# Câu 2: Kiểm định giả thiết “Hệ số hồi quy của X trong hàm hồi quy tổng thể bằng 0 với mức ý nghĩa 2%” và nêu ý nghĩa của kết quả?
import statsmodels.api as sm
import numpy as np

# Tạo dữ liệu mẫu
X = np.array([7, 6.5, 6.5, 6.0, 6.0, 6.0, 5.5, 5.5, 5.0, 4.5])
y = np.array([29, 32, 31, 34, 32, 35, 40, 43, 48, 50])

# Thêm cột giá trị 1 vào ma trận X để tính toán hệ số chặn
X = sm.add_constant(X)

# Tạo mô hình hồi quy tuyến tính
model = sm.OLS(y, X)

# Fit mô hình
results = model.fit()

results.summary()

# Kiểm định giả thiết
t_statistic = results.tvalues[1]  # Lấy giá trị t-statistic của hệ số X
p_value = results.pvalues[1]  # Lấy giá trị p-value của hệ số X

print("Giá trị t-statistic:", t_statistic)
print("Giá trị p-value:", p_value)
```
