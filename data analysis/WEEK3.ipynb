D = { H, T, H, T, H, H, T, H, H, H}.
gọi xác suất xuất hiện mặt H là theta => xác suất hiện mặt T là 1 - theta.

P(A, B) = P(A).P(B).

# Phương pháp maximun likelihood

Mục tiêu tìm tham số để likelihood trong tính toán:
Khi thực nghiệm quá nhiều thì xu hướng dẫn đến suy biến tính toán: giải pháp chuyển P(D, h) => log(P(D, h)).
P(Vỏ \ Độc tố): Vỏ variable, Độc tố condition, case.

Để tránh trường hợp xác suất 0 trong likelihood thì chúng ta sử dụng phương pháp adding one.
- Maximun likelihood.
- Adding one (MAP).
- Phân phối Markov:
  + One gram (Unitgram).
  + Two gram (Bigram).
  + Three gram (Trigram).
